```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.api import ExponentialSmoothing,SimpleExpSmoothing
```

```python
df=pd.read_csv("Electric_Production.csv")
df.head()
```

```python
df.shape
```

```python
#Moving average
plt.plot(df['Value'])
```

```python
#Moving average
plt.plot(df[1:50]['Value'])
```

```python
ma_data=df[1:50].rolling(window=10)
type(ma_data)
```

```python
ma_data_mean=ma_data.mean()
ma_data_mean.head(10)
```

```python
plt.plot(ma_data_mean)
```

```python
ma_data=df.rolling(window=10)
ma_data_mean=ma_data.mean()
plt.plot(ma_data_mean)
```

```python
plt.plot(df['Value'])
```

```python
#simkple exponential smoothing
fit1 = SimpleExpSmoothing(df['Value']).fit(smoothing_level=0.8)
fit2 = SimpleExpSmoothing(df['Value']).fit(smoothing_level=0.2)
```

```python
fit1
```

```python
plt.plot(fit1.fittedvalues)
```

```python
plt.plot(fit2.fittedvalues)
```
